# Menangani value missing

```python
df.isna().sum()#mengecek data values missing
```

menunjukan apakah dalam suatu dataframe terdapat nilai kosong, 
seperti contoh terdapat nilai kosong sebanyak 386 dalam kolom Type 2

```python
df['Type 2'].value_counts()
```

```python
df['Type 1'].value_counts()
```

Mencari tahu nilai dengan jumlah nilai pada suatu kolom

```python
mode=df['Type 2'].mode()[0]
```

Sintaks diatas adalah salah satu cara dalam menangani value missing pada kolom yang nilai valuenya bukan Integer melainkan Object/String,"Mode" adalah nilai modus yang akan digunakan untuk mengisi nilai kosong/NaN/null, akan tetapi ini masih diproses dalam proses program dibelakang, apabila kita run dan kita tampilkan maka nilai Nan/kosong masih exist

```python
df
```

karena sudah mendeklarasikan rumus mode untuk menangani missing value makas hasil proses tadi tersimpan dalam mode belum dilakukan inplace

```python
print(mode)
```

mencari tahu nilai modus dalam mode yg sudah dideklarasikan

```python
df['Type 2'].fillna(mode,inplace=True)
```

berikut ini memasukan mode yang tadi sudah diproses kedalam dataframe dengan nilai True

```python
df
```

# Manipulasi

```python
Pokemon=df.copy()#bagian manipulasi
```

di bagian ini membuat copy atau backup dataframe sebelumnya untuk dilakukan manipulasi dataframe

```python
df.shape
```

mencari tahu jumalh baris dan kolom

```python
Pokemon['Fatality']= df.Attack + df.Speed`
```

Disini membuat kolom baru dengan nama Fatality yang didalamnya terdapat nilai dari masing-masing Pokemon dengan cara menambahkan setiap value Attack dengan Speed

```python
Pokemon
```

```python
Pokemon.shape
```

```python
Pokemon['Fatality'].value_counts()
```

```python
pemukul=df[df['Attack']>=100] #query data dengan nilai attack lebih dari 100
```

Menegolompokan data yang dimana nilai attack lebih nari 100 dengan membuat variabel pemukul terlebih dahulu

```python
pemukul
```

```python
df[
    (df['Type 1']=='Water')&
    (df['Attack']>=95)
].groupby(['Name','Speed'])['Defense'].count()
```

Sitaks diatas mealkukan groupby yang artinya menampilkan data dengan cara dibentuk dengan list. data diatas adalah dimana ingin menampilkan pokemon jenis air yang nilai attacknya lebih dari 95, dan ditampilkan dalam nama pokemon dan nilai speednya
